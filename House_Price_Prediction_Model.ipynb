```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style = "whitegrid")
df = pd.read_csv(r"C:\Users\DURGA PRASAD\Downloads\house-prices-advanced-regression-techniques\train.csv")
df.head()

```

```python
print("Dataset Shape:", df.shape)
df.info()
df.describe()
```

```python
# Distribution of target variable
plt.figure(figsize=(8, 5))
sns.histplot(df['SalePrice'], kde=True) #histogram with kde curve
plt.title('Distribution of House Sale Prices')
plt.xlabel('Sale Price')

```

```python
missing = df.isnull().sum() #no.of missing values in a coloum
missing = missing[missing > 0].sort_values(ascending=False) #filters coloums with atleast 1 missing value in descending order
missing #shows the values

```

```python
missing_data = df.isnull().sum().sort_values(ascending = False)
missing_data[missing_data > 0].head(10)
```

```python
df = df.drop(['PoolQC','MiscFeature', 'Alley', 'Fence', 'MasVnrType', 'FireplaceQu'], axis=1, errors='ignore') # Drop columns with too many missing values
# Fill categorical features with 'None'
cat_cols_to_fill = ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',
                    'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
                    'MasVnrType', 'Electrical']
cat_cols_to_fill = [col for col in cat_cols_to_fill if col in df.columns]
for col in cat_cols_to_fill:
    df[col] = df[col].fillna('None')
# Fill numerical columns with median
num_cols_to_fill = ['GarageYrBlt', 'MasVnrArea', 'LotFrontage']
num_cols_to_fill = [col for col in num_cols_to_fill if col in df.columns]

for col in num_cols_to_fill:
    df[col] = df[col].fillna(df[col].median())

# Confirm there are no more missing values
df.isnull().sum().sum()
```

```python
#coverting categorical values into binary
#Get All Categorical Columns
categorical_cols = df.select_dtypes(include='object').columns
#Apply One-Hot Encoding with pd.get_dummies()
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
#check's result
df_encoded.head()
df_encoded.shape
```

```python
#Split Into Features (X) and Target (y)
y = df_encoded['SalePrice']
x = df_encoded.drop('SalePrice', axis=1)
print("xshape:", x.shape)
print("yshape:", y.shape)
```

```python
#Code to Split the Data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=42)
print("X_train shape:", x_train.shape)
print("X_test shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
#Train a House Price Prediction Model (Linear Regression)
from sklearn.linear_model import LinearRegression
model = LinearRegression()# creating model
model.fit(x_train, y_train)# traing model
```

```python
#Make Predictions
y_pred = model.predict(x_test)
print(y_pred)
```

```python
#To compare actual and predicted values
pd.DataFrame({'Actual': y_test, 'Predicted': y_pred}).head(10)
```

```python
#evaluating the model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R2: {r2}")
```

```python
#Train a House Price Prediction Model (Random Forest Regression)
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state = 42, n_estimators = 100)
model.fit(x_train, y_train)
```

```python
#Make Predictions
y_pred = model.predict(x_test)
print(y_pred)
```

```python
#To compare actual and predicted values
pd.DataFrame({'Actual': y_test, 'Predicted': y_pred}).head(10)
```

```python
#evaluating the model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R2: {r2}")
```

```python
#Visualization – Predictions vs Actuals
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.6)
plt.xlabel("Actual SalePrice")
plt.ylabel("Predicted SalePrice")
plt.title("Actual vs Predicted Sale Prices")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # ideal line
plt.grid(True)
plt.show()
```
